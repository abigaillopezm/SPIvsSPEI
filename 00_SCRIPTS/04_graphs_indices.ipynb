```python
import matplotlib.pyplot as plt
import pandas as pd
import os
```

##### Para ventana de 3 meses
Se seleccionó una estación por cada región de Jalisco, considerando:
1. Norte
2. Altos Norte
3. Altos sur
4. Ciénega
5. Sureste
6. Sur
7. Sierra de amula
8. Costa Sur
9. Costa norte
10. Sierra Occidental
11. Valles
12. Centro

```python
def graficar_estacion(id_estacion, ventana=3):
    # 1. Buscar el archivo en tu carpeta de resultados
    ruta = f"../02_OUTPUTS/ven3/RESULTADOS_V{ventana}_SMN_{id_estacion}_1991_2020.csv"
    
    if not os.path.exists(ruta):
        print(f"No se encontró el archivo para la estación {id_estacion}")
        return

    # 2. Cargar datos
    df = pd.read_csv(ruta, index_col='FECHA', parse_dates=True)
    
    # 3. Configurar la figura
    fig, ax = plt.subplots(figsize=(14, 6))
    
    # Graficar SPEI (Áreas)
    # Rellenamos de rojo cuando es sequía (menor a 0) y azul cuando es humedad (mayor a 0)
    ax.fill_between(df.index, 0, df['SPEI'], where=(df['SPEI'] >= 0), color='skyblue', alpha=0.8, label='Humedad (SPEI)')
    ax.fill_between(df.index, 0, df['SPEI'], where=(df['SPEI'] < 0), color='salmon', alpha=0.8, label='Sequía (SPEI)')
    
    # Graficar SPI (Línea para contrastar)
    ax.plot(df.index, df['SPI'], color='darkblue', linewidth=1, alpha=0.6, label=f'SPI-{ventana}')

    # 4. Detalles institucionales
    ax.axhline(0, color='black', lw=1)
    ax.axhline(-1, color='red', linestyle='--', alpha=0.3) # Umbral sequía moderada
    ax.axhline(-2, color='darkred', linestyle='--', alpha=0.5) # Umbral sequía extrema
    
    ax.set_title(f'Comparativa de Sequía Estación {id_estacion} (Escala {ventana} meses)', fontsize=14)
    ax.set_ylabel('Índice Estandarizado')
    ax.set_xlabel('Año')
    ax.legend(loc='upper right')
    ax.grid(axis='y', alpha=0.3)
    
    plt.tight_layout()
    plt.savefig(f'../02_OUTPUTS/SPEIvSPI_ven03_{id_estacion}', dpi=300, bbox_inches='tight')
    plt.show()

# --- PRUEBA CON TU ESTACIÓN ---
estacs = ("14023", "14084", "14076", "14075", "14040", "14348", "14078", "14028", "14297", "14009", "14066")


for id_estacion in estacs:
    graficar_estacion(id_estacion,ventana=3 )


# 📍Estaciones consideradas
# 1. Norte --------------------- Bolaños                    14023
# 2. Altos Norte --------------- Lagos de moreno            14084
# 3. Altos sur ----------------- Jose María                 14076
# 4. Ciénega ------------------- Jamay                      14075
# 5. Sureste ------------------- Chapala                    14040
# 6. Sur ----------------------- Jilotlán de los Dolores    14348
# 7. Sierra de amula ----------- Juchitlán                  14078
# 8. Costa Sur ----------------- Cihuatlán                  14028
# 9. Cost. N, Sierr Occ. ------- Guachinango                14297
# 10. Valles ------------------- Ameca                      14009
# 11. Centro ------------------- Guadalajara                14066
#graficar_estacion(14002, ventana=3)
```
